# Anatomy Generation Example

This notebook demonstrates how to use the new object-oriented anatomy module to generate and visualize plant cross-sections.

```python
import sys
import os
import matplotlib.pyplot as plt

# Add parent directory to path to allow importing anatomy package
# sys.path.append(os.path.abspath('..'))

from granap.needle_class import NeedleAnatomy
from granap.root_class import RootAnatomy
from granap.visualization import plot_layers_simple, plot_section
```

## Needle Anatomy

Generate a needle cross-section.

```python
# Initialize needle anatomy
needle = NeedleAnatomy()

# View listed layers
print("Default Layers:", needle.list_layers())
```

```python
# Plot layer boundaries
fig_layers = needle.plot_layers()
```

```python
# Generate and plot cells
# This performs Voronoi tessellation to fill the layers with cells
fig_cells = needle.plot_cells()
```

## Root Anatomy

Generate a root cross-section.

```python
# Initialize root anatomy
root = RootAnatomy()

# Plot layer boundaries
fig_root_layers = root.plot_layers()
```

```python
# Generate and plot cells
fig_root_cells = root.plot_cells()
```

```python
# Get statistics
stats = needle.get_statistics()
for key, value in stats.items():
    print(f"{key}: {value}")
```
